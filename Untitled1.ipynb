# 1. Load of modules

```python
import os
import re
import zipfile
import random
import urllib.request
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import (
    Input, Embedding, Conv1D, GlobalMaxPooling1D,
    Concatenate, Dropout, Dense
)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping

print("TensorFlow version:", tf.__version__)


```


# ============================================================
# 2. Reproducibility
# ============================================================


```python
SEED = 42

random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

os.environ["PYTHONHASHSEED"] = str(SEED)


```

# ============================================================
# 3. Download of the MR dataset (rt-polarity)
# ============================================================


```python
data_dir = Path("/content/mr_data")
data_dir.mkdir(parents=True, exist_ok=True)

pos_url = "https://raw.githubusercontent.com/yoonkim/CNN_sentence/master/rt-polarity.pos"
neg_url = "https://raw.githubusercontent.com/yoonkim/CNN_sentence/master/rt-polarity.neg"

pos_path = data_dir / "rt-polarity.pos"
neg_path = data_dir / "rt-polarity.neg"

if not pos_path.exists():
    urllib.request.urlretrieve(pos_url, pos_path)

if not neg_path.exists():
    urllib.request.urlretrieve(neg_url, neg_path)

print("Files downloaded:")
print(pos_path)
print(neg_path)


```

# ============================================================
# 4. Load and clean text
# ============================================================


```python
def clean_str(text: str) -> str:
    text = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", text)
    text = re.sub(r"\'s", " \'s", text)
    text = re.sub(r"\'ve", " \'ve", text)
    text = re.sub(r"n\'t", " n\'t", text)
    text = re.sub(r"\'re", " \'re", text)
    text = re.sub(r"\'d", " \'d", text)
    text = re.sub(r"\'ll", " \'ll", text)
    text = re.sub(r",", " , ", text)
    text = re.sub(r"!", " ! ", text)
    text = re.sub(r"\(", " ( ", text)
    text = re.sub(r"\)", " ) ", text)
    text = re.sub(r"\?", " ? ", text)
    text = re.sub(r"\s{2,}", " ", text)
    return text.strip().lower()

def load_mr_data(pos_file, neg_file):
    texts = []
    labels = []

    with open(pos_file, "r", encoding="latin-1") as f:
        for line in f:
            line = clean_str(line.strip())
            if line:
                texts.append(line)
                labels.append(1)

    with open(neg_file, "r", encoding="latin-1") as f:
        for line in f:
            line = clean_str(line.strip())
            if line:
                texts.append(line)
                labels.append(0)

    return texts, np.array(labels)

texts, labels = load_mr_data(pos_path, neg_path)

print("Number of samples:", len(texts))
print("Positive samples:", int(labels.sum()))
print("Negative samples:", int((labels == 0).sum()))
print("\nSample text:\n", texts[0])


```

# ============================================================
# 5. Tokenization and Padding
# ============================================================


```python
max_words = 20000
max_len = 56   # common practical choice for MR-style short sentences
embedding_dim = 100

tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
X = pad_sequences(sequences, maxlen=max_len, padding="post", truncating="post")
y = labels

word_index = tokenizer.word_index
vocab_size = min(max_words, len(word_index) + 1)

print("Vocabulary size:", vocab_size)
print("X shape:", X.shape)


```

# ============================================================
# 6. Download Glove
# ============================================================


```python
glove_dir = Path("/content/glove")
glove_dir.mkdir(parents=True, exist_ok=True)

glove_zip = glove_dir / "glove.6B.zip"
glove_txt = glove_dir / f"glove.6B.{embedding_dim}d.txt"

if not glove_txt.exists():
    if not glove_zip.exists():
        urllib.request.urlretrieve(
            "https://nlp.stanford.edu/data/glove.6B.zip",
            glove_zip
        )
    with zipfile.ZipFile(glove_zip, "r") as zf:
        zf.extract(f"glove.6B.{embedding_dim}d.txt", glove_dir)

print("GloVe file ready:", glove_txt)


```

# ============================================================
# 7. Build Embedding Matrix
# ============================================================


```python
def load_glove_embeddings(glove_file):
    embeddings_index = {}
    with open(glove_file, "r", encoding="utf-8") as f:
        for line in f:
            values = line.rstrip().split(" ")
            word = values[0]
            coefs = np.asarray(values[1:], dtype="float32")
            embeddings_index[word] = coefs
    return embeddings_index

embeddings_index = load_glove_embeddings(glove_txt)
print("Loaded embeddings:", len(embeddings_index))

embedding_matrix = np.random.normal(scale=0.6, size=(vocab_size, embedding_dim)).astype("float32")
embedding_matrix[0] = np.zeros((embedding_dim,), dtype="float32")  # padding token

hits, misses = 0, 0
for word, i in word_index.items():
    if i >= vocab_size:
        continue
    vector = embeddings_index.get(word)
    if vector is not None:
        embedding_matrix[i] = vector
        hits += 1
    else:
        misses += 1

print(f"Embedding coverage -> hits: {hits}, misses: {misses}")


```

# ============================================================
# 8. Model Definition
# ============================================================


```python
def build_kim_cnn(
    vocab_size,
    max_len,
    embedding_dim,
    embedding_matrix=None,
    trainable_embedding=True,
    num_filters=100,
    kernel_sizes=(3, 4, 5),
    dropout_rate=0.5
):
    inputs = Input(shape=(max_len,), name="input_ids")

    if embedding_matrix is not None:
        x = Embedding(
            input_dim=vocab_size,
            output_dim=embedding_dim,
            weights=[embedding_matrix],
            input_length=max_len,
            trainable=trainable_embedding,
            name="embedding"
        )(inputs)
    else:
        x = Embedding(
            input_dim=vocab_size,
            output_dim=embedding_dim,
            input_length=max_len,
            trainable=True,
            name="embedding"
        )(inputs)

    conv_blocks = []
    for k in kernel_sizes:
        conv = Conv1D(
            filters=num_filters,
            kernel_size=k,
            activation="relu",
            name=f"conv_{k}"
        )(x)
        pool = GlobalMaxPooling1D(name=f"global_maxpool_{k}")(conv)
        conv_blocks.append(pool)

    if len(conv_blocks) > 1:
        z = Concatenate(name="concat")(conv_blocks)
    else:
        z = conv_blocks[0]

    z = Dropout(dropout_rate, name="dropout")(z)
    outputs = Dense(1, activation="sigmoid", name="classifier")(z)

    model = Model(inputs=inputs, outputs=outputs, name="kim_cnn")
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

# quick sanity check
tmp_model = build_kim_cnn(
    vocab_size=vocab_size,
    max_len=max_len,
    embedding_dim=embedding_dim,
    embedding_matrix=embedding_matrix,
    trainable_embedding=False
)
tmp_model.summary()


```

# ============================================================
# 9. Train One Fold
# ============================================================


```python
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.models import load_model

def train_one_fold(
    X_train, y_train, X_val, y_val,
    model_type="rand",
    batch_size=50,
    epochs=15
):
    tf.keras.backend.clear_session()

    if model_type == "rand":
        model = build_kim_cnn(
            vocab_size=vocab_size,
            max_len=max_len,
            embedding_dim=embedding_dim,
            embedding_matrix=None,
            trainable_embedding=True
        )
    elif model_type == "static":
        model = build_kim_cnn(
            vocab_size=vocab_size,
            max_len=max_len,
            embedding_dim=embedding_dim,
            embedding_matrix=embedding_matrix,
            trainable_embedding=False
        )
    elif model_type == "nonstatic":
        model = build_kim_cnn(
            vocab_size=vocab_size,
            max_len=max_len,
            embedding_dim=embedding_dim,
            embedding_matrix=embedding_matrix,
            trainable_embedding=True
        )
    else:
        raise ValueError("model_type must be one of: rand, static, nonstatic")

    model_path = f"/content/best_model_{model_type}.keras"

    checkpoint = ModelCheckpoint(
        model_path,
        monitor="val_accuracy",
        mode="max",
        save_best_only=True,
        verbose=0
    )

    callbacks = [
        EarlyStopping(
            monitor="val_accuracy",
            patience=3,
            mode="max",
            restore_best_weights=True
        ),
        checkpoint
    ]

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        verbose=0,
        callbacks=callbacks
    )

    #  Assures that the best model is used
    model = load_model(model_path)

    y_prob = model.predict(X_val, verbose=0).ravel()
    y_pred = (y_prob >= 0.5).astype(int)
    acc = accuracy_score(y_val, y_pred)

    return model, history.history, acc, y_pred, y_prob


```

# ============================================================
# 10. Cross-Validation Experiments
# ============================================================


```python
def run_cross_validation(X, y, model_type="rand", n_splits=10):
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=SEED)

    fold_accuracies = []
    histories = []
    all_true = []
    all_pred = []

    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), start=1):
        print(f"\n=== {model_type.upper()} | Fold {fold}/{n_splits} ===")

        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]

        model, history, acc, y_pred, y_prob = train_one_fold(
            X_train, y_train, X_val, y_val, model_type=model_type
        )

        fold_accuracies.append(acc)
        histories.append(history)
        all_true.extend(y_val.tolist())
        all_pred.extend(y_pred.tolist())

        print(f"Fold accuracy: {acc:.4f}")

    results = {
        "model_type": model_type,
        "fold_accuracies": fold_accuracies,
        "mean_accuracy": float(np.mean(fold_accuracies)),
        "std_accuracy": float(np.std(fold_accuracies)),
        "histories": histories,
        "all_true": np.array(all_true),
        "all_pred": np.array(all_pred)
    }
    return results

results_rand = run_cross_validation(X, y, model_type="rand", n_splits=10)
results_static = run_cross_validation(X, y, model_type="static", n_splits=10)
results_nonstatic = run_cross_validation(X, y, model_type="nonstatic", n_splits=10)


```

# ============================================================
# 13. Classification Reports
# ============================================================


```python
def show_metrics(results, model_name):
    print(f"\n===== {model_name} =====")
    print(classification_report(results["all_true"], results["all_pred"], digits=4))

    cm = confusion_matrix(results["all_true"], results["all_pred"])
    print("Confusion Matrix:\n", cm)

show_metrics(results_rand, "CNN-rand")
show_metrics(results_static, "CNN-static")
show_metrics(results_nonstatic, "CNN-nonstatic")


```

```python
def plot_history(history_dict, title_prefix=""):
    plt.figure(figsize=(8, 5))
    plt.plot(history_dict["accuracy"], label="train_accuracy")
    plt.plot(history_dict["val_accuracy"], label="val_accuracy")
    plt.title(f"{title_prefix} Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

plot_history(results_rand["histories"][0], "CNN-rand")
plot_history(results_static["histories"][0], "CNN-static")
plot_history(results_nonstatic["histories"][0], "CNN-nonstatic")


```
